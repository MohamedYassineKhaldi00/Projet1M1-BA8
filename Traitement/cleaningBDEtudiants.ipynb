```python
# fill missing burth places with Tunis, the most recurrent place across students.
import statistics
df['Birth place'] = df['Birth place'].fillna(statistics.mode(df['Birth place']))
```

```python
df.isna().sum()
```

```python
#fill highschool graduation date based on the assumption that the student never failed any class since graduation. Example: If current class is M1, meaning current level is 4, we assum that he graduated from highschool 4 years prior to when current level school year ends: 2025 - 4
df['Highschool graduation date'] = df['Highschool graduation date'].fillna(2025 - df['Class level'])
```

```python
df[(df['Region'].isna())& (~df['Birth place'].isin(df['Region'].unique())) & (df['Nationality'] != 'TUNISIENNE')]
```

```python
# Fill region values with a series of conditions
# 1. If Birth place is in existing regions, use Birth place
# 2. If Tunisian and Birth place not in regions, use 'Tunis'
# 3. If not Tunisian, use 'Foreign'

def fill_region(row):
    if pd.isna(row['Region']):
        if row['Birth place'] in df['Region'].unique():
            return row['Birth place']
        elif row['Nationality'] == 'TUNISIENNE':
            return 'Tunis'
        else:
            return 'Foreign'
    return row['Region']

df['Region'] = df.apply(fill_region, axis=1)
```

```python
# We assumer that students with missing highschool values went to highschools in their regions:

df['Highschool'] = df['Highschool'].fillna('Lycee ' + df['Region'])
```

```python
# missing values in the Studies level column are filled with the most common value within each Degree group. If no common value exists, the missing value is nan.
df['Studies level'] = df.groupby('Degree')['Studies level'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan))
```

```python
df['Studies level'].isna().sum()
```

##### "ESB entry date" column, the number of null values is the most significant (1233). We will fill missing values based on the logic that a bachelor student enters ESB the same year they graduated highschool. Same logic is applied for first year master students, with the condition that 'Previous year result' column's value includes 'Admis', therefore minimizing inaccuracy. For second year master students, we will consider they entered ESB at their first master year.

```python
# Fill missing entry date values for bachelor students
mask_licence = (
    (df['ESB entry date'].isna()) & (df['Class level'].isin([1, 2, 3])) & (~df['Previous year result'].str.contains('Redouble', na=False)) )

df.loc[mask_licence, 'ESB entry date'] = df['Highschool graduation date']

# Fill missing entry date values for master students
mask_master = (
    (df['ESB entry date'].isna()) & df['Class level'].isin([4, 5]) & (~df['Previous year result'].str.contains('Redouble', na=False)) )

df.loc[mask_master, 'ESB entry date'] = df.loc[mask_master, 'Class level'].map({4: 2021, 5: 2023})


# For remaining missing values, use the median entry date for the same degree
df['ESB entry date'] = df.groupby('Degree')['ESB entry date'].transform(lambda x: x.fillna(x.median()))
```

```python
#We fill missing bac type values with the most recurrent bac type in the degree program the student is enrolled in
df['Bac type'] = df['Bac type'].fillna(df.groupby('Degree')['Bac type'].transform(lambda x: x.mode().iloc[0]))
```

```python
df.groupby('Bac type')['Highschool graduation grade'].median()
```

```python
# filling missing values in the 'Highschool graduation grade' column using the median grades of students that correspond with the same Bac type. Example: A Bac Math student will have his missing highschool graduation grade filled by the median of grades of all students holding a bac math.
df['Highschool graduation grade'] = df['Highschool graduation grade'].fillna(df.groupby('Bac type')['Highschool graduation grade'].transform('median'))
```

```python
# Wwe fill previous class empty fields by substracting 1 from the current class digit indicating the level. Example : Current class 2-BA-8. Assuming that student remains in the same group, we substract 1 from the first digit (2). PRevious class becomes therefroe : 1-BA-8
def fill_previous_class(row):
    if pd.isna(row['Previous class']):
        current_class = row['Current class']
        
        # Ensure the first character is numerical and properly subtract 1
        if current_class and current_class[0].isdigit():
            if int(current_class[0]) > 1:
                # Subtract 1 only from the first digit of the current class and attach rest of value
                previous_class = str(int(current_class[0]) - 1) + current_class[1:]
            else:
                # If the first digit is 1, return 'Unknown'
                previous_class = 'External'
        else:
            # Handle unexpected values safely
            previous_class = 'External'
        
        return previous_class
    # If the 'Previous class' already has a value, keep as is
    return row['Previous class']

df['Previous class'] = df.apply(fill_previous_class, axis=1)
```
